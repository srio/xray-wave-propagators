Implementing the FrFT based 1D wave propagation as per "Discrete fractional Fourier transform as a fast algorithm for evaluating the diffraction pattern of pulsed radiation" by Hanna et.al., J. Opt. Soc. Am. A / Vol. 28, No. 8 / August 2011

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import gaussian
from tqdm import tqdm
```

```python
N = 200
f = gaussian(N,10)
x = np.linspace(-1,1,N)
plt.plot(x,f)
plt.ylabel('f')
plt.xlabel('x')
plt.show()
```
