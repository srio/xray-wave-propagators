```python
import numpy as np
import matplotlib.pyplot as plt
import numexpr as ne
```

```python
energy = 1000
wavel = (1240/energy)*10**(-9)
z = 5e-6
N = 5000
xlim = 2.5e-6
domain = np.linspace(-xlim,xlim,N)
step = 2*xlim/N
in_wave = np.zeros(N)
in_wave[int(N/2)-int(N/8):int(N/2)+int(N/8)] = 1
out_wave = np.zeros((N),dtype='complex128')
pi = np.pi
```

```python
# 1D kernel to integrate
def integrate(x1,loc,dist,wavel):
    x = domain[loc]
    f = in_wave[loc]
    return f*np.exp((-1j*pi*x*x)/(wavel*dist))*np.exp((-1j*2*pi*x*x1)/(wavel*dist))
```

```python
for i in range(N):
    for j in range(N):
        x = domain[j]
        f = in_wave[j]
        out_wave[i] += integrate(domain[i],j,z,wavel)
out_wave = out_wave*1/np.sqrt(1j*wavel*z)*step
```

```python
f, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(domain*1e6,np.abs(in_wave))
ax1.set_title('Input', fontsize = 15)
ax2.plot(domain*1e6,np.abs(out_wave),'b')
#ax2.plot(domain*1e6,np.angle(o),'g')
ax2.set_title('Output', fontsize = 15)
f.set_size_inches(20, 10, forward=True)
f.suptitle('Fresnel Number : '+str((xlim**2)/(wavel*z)),fontsize = 25)
plt.show()
```
